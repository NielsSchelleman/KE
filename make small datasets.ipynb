```python
import pandas as pd
```

```python
long_lat = pd.read_csv("Train_Crossing_Lon_Lat.csv")
car_acc = pd.read_csv('US_Accidents_Dec21_updated.csv')
```

```python
car_acc.columns
```

```python
car_acc['index_Time'] = pd.to_datetime(car_acc['Start_Time'],yearfirst=True)
car_acc = car_acc[['index_Time','Severity','Start_Lat','Start_Lng','Distance(mi)','Number','City','County','State','Temperature(F)',
                  'Wind_Chill(F)','Humidity(%)','Pressure(in)','Visibility(mi)','Wind_Direction','Wind_Speed(mph)','Precipitation(in)', 'Weather_Condition',
                  'Bump', 'Crossing', 'Give_Way', 'Junction', 'No_Exit', 'Railway','Roundabout', 'Station', 'Stop', 'Traffic_Calming', 'Traffic_Signal','Turning_Loop', 'Sunrise_Sunset', 'Civil_Twilight', 'Nautical_Twilight','Astronomical_Twilight']]
```

```python
car_acc = car_acc[car_acc['Railway']==True]
```

```python
long_lat.loc[long_lat['Day']==0,'Day'] = '1'
long_lat.loc[long_lat['Incident Month']==0,'Incident Month'] = '1'
long_lat[long_lat['Time'].str.contains('None')] = '0:00'
thing = long_lat['Incident Year'].astype(str)+'/' + long_lat['Incident Month'].astype(str)+'/' +long_lat['Day'].astype(str) +' ' + long_lat['Time']
long_lat['index_Time'] = pd.to_datetime(thing, yearfirst=True)
```

```python
long_lat = long_lat[['index_Time','Maintenance Incident Year',
 'Maintenance Incident Month','Estimated Vehicle Speed','Vehicle Direction','Highway User Position', 'Railroad Car Unit Position','Equipment Type','Track Type','Track Class','Number of Locomotive Units',
 'Number of Cars','Train Speed','Train Direction','Signaled Crossing Warning', 'Roadway Condition','Crossing Warning Location', 'Crossing Illuminated','User Age', 'User Gender','Highway User Action','View Obstruction','Vehicle Damage Cost','Number Vehicle Occupants','Latitude',
 'Longitude']]
```

```python
list(long_lat['Latitude'].unique())
```

```python
long_lat = long_lat[long_lat['Latitude'] != '0:00']
long_lat['Latitude'] = long_lat['Latitude'].astype(str).str.strip('°').astype(float)
long_lat['Longitude'] = long_lat['Longitude'].astype(str).str.strip('°').astype(float)
long_lat = long_lat[long_lat['Latitude']>25.84]
long_lat = long_lat[long_lat['Latitude']<49.38]
long_lat = long_lat[long_lat['Longitude']>-130]
long_lat = long_lat[long_lat['Longitude']<-65]
long_lat.dropna(subset=['Longitude','Latitude','index_Time'], inplace=True)
long_lat
```

```python
car_acc.dropna(subset=['Start_Lat','Start_Lng','index_Time'], inplace=True)
```

```python
import geopy.distance as gd
```

```python
indices = []
x = 0
for i, (time, lat, long) in car_acc[['index_Time','Start_Lat','Start_Lng']].iterrows():
    x+=1
    if x%1 == 0:
        print(x)
    sub_long_lat  = long_lat[long_lat['index_Time']-time < pd.Timedelta(days=1)]    
    for j,(lat2, long2) in sub_long_lat[['Latitude','Longitude']].iterrows():
        if gd.geodesic((lat,long),(lat2,long2)).km < 2:
            indices.append((i,j))
```

```python
indices = []
x=0
for i, (time, lat, long) in long_lat[['index_Time','Latitude','Longitude']].iterrows():
    x+=1
    if x%1000 == 0:
        print(x)
    sub_acc = car_acc[car_acc['index_Time']-time < pd.Timedelta(days=1)]
    sub2_acc = sub_acc[abs(sub_acc['Start_Lat']-lat)<0.5]
    sub2_acc = sub2_acc[abs(sub2_acc['Start_Lng']-long)<0.5]
    for j, (lat2, long2) in sub2_acc[['Start_Lat','Start_Lng']].iterrows():
        if gd.geodesic((lat,long),(lat2,long2)).km < 1:
            indices.append((i,j))
    
```

```python
import numpy as np
```

```python
indexed_car_acc = car_acc[car_acc.index.isin(np.array(indices)[:,1])]
indexed_train = long_lat[long_lat.index.isin(np.array(indices)[:,0])]
```
